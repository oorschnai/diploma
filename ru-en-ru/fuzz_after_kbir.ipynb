```python
import pandas as pd
import numpy as np
import json
from thefuzz import fuzz
from thefuzz import process
```

```python
argos_file = 'keyph20ru.json'
marian_file = 'marian10ru.json'


f = open('rpd_data.json')
rpd_data = json.load(f)

f.close()

f = open(argos_file)
argos_KW = json.load(f)

f.close()

f = open(marian_file)
marian_KW = json.load(f)

f.close()
```

```python
marian_KW
```

```python
argos_res = []

for i in range(len(argos_KW)):
    text = rpd_data[i]['text']
    arr = []
    for word in argos_KW[i]:
        number = fuzz.partial_ratio(word, text)
        for sentence in text.split('.'):
            if (fuzz.partial_ratio(word, sentence)) > number:
                number = fuzz.partial_ratio(word, sentence)
                
        arr.append(number)
                
    argos_res.append(arr)
```

```python
result = []

for j in range(len(argos_res)):
    print(j)
    arr = []
    for i in range(len(argos_res[j])):
        if argos_res[j][i] >= 90:
            print("________________", argos_KW[j][i], "|", argos_res[j][i], "________________")
            arr.append(argos_KW[j][i])
        else:
            print(argos_res[j][i], "|", argos_KW[j][i])
    result.append(arr)
    print("////////////////////////////////////////////////////////")
```

```python
with open('argos_20_after_fuzz.json', 'w') as outfile:
    outfile.write(json.dumps(result, ensure_ascii=False, indent=4))
```

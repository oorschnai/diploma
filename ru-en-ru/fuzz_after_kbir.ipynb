```python
import pandas as pd
import numpy as np
import json
from thefuzz import fuzz
from thefuzz import process
```

```python
argos_file = 'keyph20ru.json'
marian_file = 'marian10ru.json'


f = open('rpd_data.json')
rpd_data = json.load(f)

f.close()

f = open(argos_file)
argos_KW = json.load(f)

f.close()

f = open(marian_file)
marian_KW = json.load(f)

f.close()
```

```python
argos_res = []
sentences = []

for i in range(len(argos_KW)):
    text = rpd_data[i]['text']
    #print(text.split('.'))
    arr = []
    s = []
    for word in argos_KW[i]:
        number = fuzz.partial_ratio(word, text)
        s1 = '*текст*'
        #print("_________________", word, number)
        for sentence in text.split('.'):
            if len(word) > len(sentence):
                continue
            if (fuzz.partial_ratio(word, sentence)) >= number:
                number = fuzz.partial_ratio(word, sentence)
                #print(number, word, '\n', sentence)
                s1 = sentence
                
        arr.append(number)
        s.append(s1)
               
    sentences.append(s)
    argos_res.append(arr)
```

```python
result = []
result1 = []

for j in range(len(argos_res)):
    print(j)
    arr = []
    arr1 = []
    for i in range(len(argos_res[j])):
        if argos_res[j][i] >= 90:
            print("________________", argos_KW[j][i], "|", argos_res[j][i], "________________")
            arr.append(argos_KW[j][i])
            words_arr = []
            for word in argos_KW[j][i].split(' '):
                m = 0
                chosen_w = ''
                for w in sentences[j][i].split(' '):
                    if (fuzz.ratio(w, word) > m):
                        m = fuzz.ratio(w, word)
                        chosen_w = '' + w
                print(word, "->", chosen_w)
                words_arr.append(chosen_w)
            arr1.append((' ').join(words_arr))
            print('!!!!!!!!', (' ').join(words_arr))
                
        else:
            print(argos_res[j][i], "|", argos_KW[j][i])
        print("sentence: '", sentences[j][i], "'")
    result.append(arr)
    result1.append(arr1)
    print("////////////////////////////////////////////////////////")
```

```python
pr = 0
for i in range(len(result)):
    print(i)
    proc = round(len(result[i]) / len(argos_res[i]) * 100, 3)
    print(proc, '%')
    pr += proc
    for j in range(len(result[i])):
        print(result[i][j], "->", result1[i][j])
    print('=========================================================================')
    
print('Средний процент', pr / len(result))
```

```python
with open('argos_20_after_fuzz.json', 'w') as outfile:
    outfile.write(json.dumps(result, ensure_ascii=False, indent=4))
```

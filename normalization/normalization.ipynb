```python
import csv

words_only = []

with open('words_only.csv', newline='\n') as csvfile:
    reader = csv.reader(csvfile, delimiter="'", quotechar='|')
    words_only = [row for row in reader]

#words_only
```

```python
#print(words_only)
words = []

for i in words_only:
    arr = []
    for j in i:
        if j != '' and j != ', ' and j != '[]' and j != '[' and j != ']':
            arr.append(j)
    #arr = [j  for j in i]
    words.append(arr)
    
#words
```

```python
words_no_repeat = []

for i in words:
    words_no_repeat.append(list(set(i)))
```

```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer


model_name_or_path = "sberbank-ai/rugpt3small_based_on_gpt2"
tokenizer = GPT2Tokenizer.from_pretrained(model_name_or_path)
model = GPT2LMHeadModel.from_pretrained(model_name_or_path)
```

```python
template = '''магазины приложений -> магазин приложений
человеко-машинного интерфейса -> человеко-машинный интерфейс
принципах взаимодействия нейронов -> принципы взаимодействия нейронов
отладки приложений -> отладка приложений
содержанием дисциплины -> содержание дисциплины
MATLAB -> MATLAB
крокодил -> крокодил
Javascript -> Javascript
Sjrfjdbdj -> Sjrfjdbdj
манекен -> манекен
телефоном -> телефон
спорного -> спорный
пара -> пара
болота -> болото
системный блок -> системный блок
старости -> старость
производственная практика -> производственная практика
шаблонных действий -> шаблонные действия
потраченного времени -> потраченное время
российского -> российский
смелого -> смелый
рукой -> рука
старого Java -> старый Java
собранного хвороста -> собранный хворост
запутанного дела -> запутанное дело
Daeafsdv GUsdbbjad FUdsg -> Daeafsdv GUsdbbjad FUdsg
ворона -> ворона
практических навыков разработки -> практические навыки разработки
'''

text = template + '''смелости -> '''
input_ids = tokenizer.encode(text, return_tensors="pt")
out = model.generate(input_ids, max_new_tokens=2)
generated_text = list(map(tokenizer.decode, out))[0]
print(generated_text)
```

```python
all_in_one = []

for i in words_no_repeat:
    for j in i:
        all_in_one.append(j)
        
all_in_one = list(set(all_in_one))

all_in_one
```

```python
norm2 = []

for word in all_in_one:
    text = template + word + ''' -> '''
    input_ids = tokenizer.encode(text, return_tensors="pt")
    out = model.generate(input_ids, max_new_tokens=2)
    generated_text_old = list(map(tokenizer.decode, out))[0]
    flag = 1
    counter = 0
    while(flag and counter < 5):
        counter += 1
        input_ids = tokenizer.encode(generated_text_old, return_tensors="pt")
        out = model.generate(input_ids, max_new_tokens=2)
        generated_text = list(map(tokenizer.decode, out))[0]
        print(generated_text.replace(template, ''))
        differences = generated_text.replace(generated_text_old, '')
        newl = generated_text.count('\n')
        if len(differences) < 2 or newl > new_lines_count:
            flag = 0
        generated_text_old = "" + generated_text
        
    norm2.append(generated_text.replace(text, '').replace('\n', ''))

    
```

```python
from natasha import MorphVocab

morph_vocab = MorphVocab()
```

```python
from natasha import (
    Segmenter,
    
    NewsEmbedding,
    NewsMorphTagger,
    NewsSyntaxParser,
    
    Doc
)

segmenter = Segmenter()
emb = NewsEmbedding()
morph_tagger = NewsMorphTagger(emb)
syntax_parser = NewsSyntaxParser(emb)


doc = Doc(text)

doc.segment(segmenter)
doc.tag_morph(morph_tagger)
doc.parse_syntax(syntax_parser)
```

```python
doc.tokens
```

```python
for span in doc.tokens:
    span.normalize(morph_vocab)
    
{_.text: _.normal for _ in doc.spans}
```

```python
import pandas as pd
import numpy as np
from IPython.display import display, HTML
from thefuzz import fuzz
from thefuzz import process
```

```python
normalization_output = pd.read_csv('norm_pub_large_200.csv').rename({'large_without_trash': 'norm'}, axis=1)
normalization_output
```

```python
fuzzed = []
for i in range(len(normalization_output)):
    if type(normalization_output.norm[i]) == float and np.isnan(normalization_output.norm[i]):
        fuzzed.append(0)
    else:
        fuzzed.append(fuzz.ratio(normalization_output.word[i], normalization_output.norm[i]))
        
normalization_output['fuzz'] = fuzzed
```

```python
#display(HTML(normalization_output[normalization_output.fuzz < 100].to_html()))
```

```python
normalization_output['selected'] = normalization_output[normalization_output.fuzz > 91].norm
normalization_output.loc[normalization_output.selected.isnull(), ['selected']] = normalization_output[normalization_output.selected.isnull()].word
```

```python
display(HTML(normalization_output.to_html()))
```

```python
result = normalization_output[['word', 'selected']].rename({'selected': 'result'}, axis = 1)
result['identical'] = (result.word == result.result)

result.style.background_gradient(axis=None, vmin=1, vmax=5, cmap="YlGnBu")
```
